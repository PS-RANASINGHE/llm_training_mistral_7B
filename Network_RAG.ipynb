```python
from langchain_objectbox.vectorstores import ObjectBox
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_community.document_loaders import PyPDFLoader
from langchain_core.prompts import ChatPromptTemplate, PromptTemplate
from langchain.chains import RetrievalQA
import os
import re
import time

# Set up OpenAI API key
os.environ["OPENAI_API_KEY"] = "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"

# Load and split the document
loader = PyPDFLoader('/content/extracted_data_fixed.pdf')
data = loader.load()
text_splitter = RecursiveCharacterTextSplitter()
documents = text_splitter.split_documents(data)

# Initialize embeddings and vector store with specified embedding dimensions
embeddings = OpenAIEmbeddings()
embedding_dimensions = 768  # Using specified dimension
unique_db_path = f"/content/objectbox_{int(time.time())}"
vector = ObjectBox.from_documents(documents, embedding=embeddings, embedding_dimensions=embedding_dimensions, db_directory=unique_db_path)
```

```python
# Initialize the LLM
llm = ChatOpenAI(model="gpt-3.5-turbo")

# Define ESG compliance templates for Customer Scope V1 and V2
template_version_1 = """please act as a networking engineer and analyze the given document and answer the networking related quesions asked

{context}
Network assistance question Question:
{question}
Comprehensive networking answer:"""

template_version_2 = """
please act as a networking engineer and analyze the given document and answer the networking related quesions asked

{context}
Network assistance question Question:
{question}
Comprehensive networking answer:"""

# Function to select and return the appropriate prompt template based on the version
def get_esg_prompt(version):
    if version == 1:
        return PromptTemplate(input_variables=["context", "question"], template=template_version_1)
    elif version == 2:
        return PromptTemplate(input_variables=["context", "question"], template=template_version_2)
    else:
        raise ValueError("Unsupported version. Please select version 1 or 2.")

# Function to perform ESG compliance check based on selected version
def get_networking_answer(version, question):
    # Select the appropriate prompt based on version
    selected_prompt = get_esg_prompt(version)

    # Set up the QA chain with the selected prompt
    qa_chain = RetrievalQA.from_chain_type(
        llm=llm,
        retriever=vector.as_retriever(),
        chain_type_kwargs={"prompt": selected_prompt}
    )

    # Run the compliance check
    testresult = qa_chain({"query": question})
    final = testresult['result']
    # Split the text at full stops and iterate over sentences
    for sentence in final.split('. '):  # Splitting at '. ' to avoid breaking abbreviations
        print(sentence.strip())  # Print each sentence after stripping extra spaces
```

```python
question = """
What is tosibox 175
              """
get_networking_answer(version=1, question=question)
```

```python
question = """
What are the specifications of tosibox 175
              """
get_networking_answer(version=1, question=question)
```
